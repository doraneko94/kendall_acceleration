```python
import kendall_py, time
import numpy as np
import scipy as sp
from matplotlib import pyplot as plt
```

```python
x = [6, 4, 9, 7, 3, 6, 1]
y = [2, 1, 4, 3, 2, 1, 2]
```

```python
print("tau: {:.2f}".format(kendall_py.tau(x, y)))
zb = kendall_py.zb(x, y)
print("zB: {:.2f}".format(zb))
```

```python
np.random.seed(0)
plt.figure(figsize=(4, 3))
b = plt.hist(np.random.normal(size=100000), bins=50, alpha=0.6)
plt.hist(np.zeros(int(b[0][sum(b[1] < zb) - 1])) + zb, bins=b[1], color="r", label="$z_B$={:.2f}".format(zb))
plt.xlim(-4, 4)
plt.xlabel("$z_B$")
plt.legend()
plt.tick_params(left=False, labelleft=False)
plt.tight_layout()
plt.savefig("img/distribution_zb.png")
plt.show()
```

```python
print("上位 {:.2f}%".format((1 - sp.stats.norm.cdf(zb)) * 100))
```

```python
n = 100000
np.random.seed(1)
z = np.random.random(n)
x = z + np.random.random(n) * 0.5
y = z + np.random.random(n) * 0.5
```

```python
fig, axes = plt.subplots(2, 1, figsize=(8, 2), sharex=True)
axes[0].plot(x[:500], c="C0")
axes[0].set_xlim(0, 500)
axes[0].set_ylabel("X")
axes[1].plot(y[:500], c="C1")
axes[1].set_ylabel("Y")
plt.tight_layout()
plt.savefig("img/xy.png")
```

```python
start = time.time()
tau = kendall_py.tau_raw(x, y)
print("計算時間: {:,.2f} ms".format((time.time() - start) * 1000))
print("tau: {:.5f}".format(tau))
```
